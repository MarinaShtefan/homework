Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам: оценка 2 и ниже — низкий рейтинг; оценка 4 и ниже — средний рейтинг; оценка 4.5 и 5 — высокий рейтинг.

```python
import pandas as pd
ratings = pd.read_csv('ratings.csv') 
movies = pd.read_csv('movies.csv') 

def class_film(rating):
    if rating <=2: 
        result='низкий рейтинг'
    elif rating <=4: 
        result='средний рейтинг'
    else:
        result='высокий рейтинг'
    return result
ratings['class'] = ratings['rating'].apply(class_film)
ratings = ratings.merge(movies, on='movieId', how='left')
ratings[['userId', 'title', 'rating', 'class']]
```

Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону. 
Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.
Результат классификации запишите в отдельный столбец region.

```python
import pandas as pd
keyw = pd.read_csv('keywords.csv', sep=',') 
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
def geo(keyword):
    keyword_lower = str(keyword).lower() # перевожу значение столбца в строчку и маленькими буквами
    words = keyword_lower.split() # Список слов для поиска
    for region, cities in geo_data.items():
        for city in cities:  # Проверяем каждый город в текущем регионе
            if city in words:
                return region 
    return 'undefined'
keyw['region'] = keyw['keyword'].apply(geo)
keyw
```

### Создаем словарь purchases

```python
import json

purchases = {}
with open('purchase_log.txt', 'r', encoding='utf-8') as file:
       for line in file:
        record = json.loads(line)
        user_id = record["user_id"]
        category = record["category"]
        purchases[user_id] = category
        
```

### Создаем файл funnel.csv и загружаем в него данные

```python
with open('funnel.csv', 'w', encoding='utf-8') as csv_file:
    with open('visit_log.csv', 'r', encoding='utf-8') as users_file:
        for line in users_file:
            columns = line.strip().split(",") # Отделяю столбцы друг от друга 
            user_id = columns[0]
            sourse = columns[1]
            
            category = purchases.get(user_id, None) # Добавляю категорию
            
            if category is not None:
                csv_file.write(f"{user_id},{sourse},{category}\n") # Записываю результат в файл
```

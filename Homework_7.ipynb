Задание 1
Печатные газеты использовали свой формат дат для каждого выпуска. Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:
The Moscow Times — Wednesday, October 2, 2002
The Guardian — Friday, 11.10.13
Daily News — Thursday, 18 August 1977

```python
import time
from datetime import date
from datetime import datetime
from datetime import datetime, timedelta
a='Wednesday, October 2, 2002'
b='Friday, 11.10.13'
c='Thursday, 18 August 1977'
date_Times=datetime.strptime(a, "%A, %B %d, %Y")
date_Guardian=datetime.strptime(b, "%A, %d.%m.%y")
date_Daily_News=datetime.strptime(c, "%A, %d %B %Y")
print(date_Times,date_Guardian,date_Daily_News)
```

Задание 2 (необязательное)
Напишите функцию date_range, которая возвращает список дат за период от start_date до end_date. Даты должны вводиться в формате yyyy-mm-dd. В случае неверного формата или при start_date > end_date должен возвращаться пустой список.

```python
import time
from datetime import date
from datetime import datetime
from datetime import datetime, timedelta
date_range=('2024-01-01','2024-01-15')
date_range_list=[]
try:
    start_date_datetime=datetime.strptime(date_range[0], '%Y-%m-%d') 
    end_date_datetime=datetime.strptime(date_range[1], '%Y-%m-%d') 
    while start_date_datetime<=end_date_datetime:
        c=start_date_datetime.strftime('%Y-%m-%d')
        date_range_list.append(c)
        start_date_datetime+=timedelta(days=1)
except ValueError:
    print()

print(date_range_list)
```

Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
movies = pd.read_csv('movies.csv') 
ratings = pd.read_csv('ratings.csv') 
ratings.head()
max_rating=ratings.query('rating == 5').groupby('movieId').count().sort_values('userId',ascending=False).head(1)
joined = max_rating.merge(movies, on='movieId', how='left')
joined.head()
```

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

```python
import pandas as pd
power = pd.read_csv('power.csv') 
pos_power=power.query('quantity > 0')
sum_power=pos_power[pos_power['country'].isin(['Latvia', 'Lithuania', 'Estonia'])]
sum_power_kat=sum_power[sum_power['category'].isin([4,12,21])]
sum_power_kat_year=sum_power_kat[sum_power_kat['year'].isin([2005,2006,2007,2008,2009,2010])]
total=sum_power_kat_year['quantity'].sum()
print(total)
```
